# Examples of Data Distribuitions

Uniform Distribuition

```python
import numpy as np 
import matplotlib.pyplot as plt 

values = np.random.uniform(-10, 10, 10000)
plt.hist(values, 50)
plt.show()
```

# Normal / Gaussian

Visualize the probability density function:

```python
gaussian = np.random.normal(0, 20, 10000)

plt.hist(gaussian, 50)
plt.show()
```
